# Data Description:

The data at hand contains medical costs of people
characterized by certain attributes.
# Domain:
Healthcare
# Context:
Leveraging customer information is paramount for most
businesses. In the case of an insurance company, attributes of
customers like the ones mentioned below can be crucial in
making business decisions. Hence, knowing to explore and
generate value out of such data can be an invaluable skill to
have.
# Attribute Information:
age : age of primary beneficiary
sex : insurance contractor gender, female, male
bmi : Body mass index, providing an understanding of body,
weights that are relatively high or low relative to height,
objective index of body weight (kg / m ^ 2) using the ratio of
height to weight, ideally 18.5 to 24.9
children : Number of children covered by health insurance /
Number of dependents
smoker : Smoking
region : the beneficiary's residential area in the US, northeast,
southeast, southwest, northwest.
charges : Individual medical costs billed by health insurance.

```python
#OutLiers in 'age'
sns.boxplot(x='age', data=df_insurance)
```

```python
#OutLiers in 'charges'
sns.boxplot(x='charges', data=df_insurance)
```

```python
#Distribution of sex
df_insurance.groupby(df_insurance['sex']).count()
```

```python
#More males compared to females.
sns.countplot(x='sex', data=df_insurance)
```

```python
#Distribution of smoker
df_insurance.groupby(df_insurance['smoker']).count()
```

```python
#No of non smokers are more. Female non smokers are more than male
sns.countplot(x='smoker',hue='sex', data=df_insurance)
```

```python
df_insurance.groupby(df_insurance['children']).count()
```

```python
#Distribution of children
sns.countplot(x='children',data=df_insurance)
```

```python
sns.countplot(x='children',hue='sex',data=df_insurance)
```

```python
sns.pairplot(df_insurance,diag_kind='kde')
```

```python
#From the pairplot it can be concluded again that the data for 'charges' is skewed.
#'bmi'has perfectly normal distribution of the data.
#The small peaks in the 'children' feature corresponds to the no of childrens
#'Age' has a spread out distribution
```

```python
sns.lmplot(x='charges', y='age', hue='smoker', data=df_insurance,scatter=True,fit_reg=False)
```

```python
#Charges of people who smoke is significatly higher than non smokers.
#Also People with age > 45 and are smoker have the highest insurance charges
```

```python
sns.barplot(x='sex', y='bmi', data=df_insurance)
```

```python
#bmi for male and female are almost the same
```
