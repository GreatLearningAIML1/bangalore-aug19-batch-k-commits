```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
dataFrame = pd.read_csv("Laliga.csv")
dataFrame.head()
```

```python
#replace '-' with 0
dataFrame.replace(to_replace='-',value=0, inplace=True)
```

```python
dataFrame.columns = dataFrame.iloc[0]
dataFrame = dataFrame[1:]
```

```python
#Print all the teams which have started playing between 1930-1980.
dataFrame.Team[(dataFrame['Debut'] >= '1930') & (dataFrame['Debut'] <= '1980')]
```

```python
#Print the list of teams which came Top 5 in terms of points
dataFrame['Points']=dataFrame['Points'].astype(int)
dataFrame.sort_values(by=['Points'],ascending=False).head()
```

```python
#Write a function with name “Goal_diff_count” which should return all the teams with their 
#Goal Differences.Using the same function, find the team which has maximum and minimum 
#goal difference.

dataFrame['GoalsFor'] = dataFrame['GoalsFor'].astype(int)
dataFrame['GoalsAgainst'] = dataFrame['GoalsAgainst'].astype(int)
def Goal_Diff_Count(value):
    value['GoalDiff']= value['GoalsFor'] - value['GoalsAgainst']
    for i in range(len(value)):
        print(value.iloc[i,1], value.iloc[i,20])
    value.sort_values(by='GoalDiff',ascending=False)
    max_value = value['GoalDiff'].max()
    min_Value = value['GoalDiff'].min()
    print("Team with maximum goal difference :",value.Team[value['GoalDiff'] == max_value])
    print("Team with minimum goal difference :",value.Team[value['GoalDiff'] == min_Value])

Goal_Diff_Count(dataFrame)
```

```python
#Create a new column with name “Winning Percent” and append it to the data set
dataFrame['GamesWon'] = dataFrame['GamesWon'].astype(int)
dataFrame['GamesPlayed'] = dataFrame['GamesPlayed'].astype(int)

dataFrame['Winning Percent'] = (dataFrame['GamesWon'] / dataFrame['GamesPlayed'])*100
```

```python
#If there are any numerical error, replace it with 0%
dataFrame.fillna(0,inplace=True)
```

```python
#Print the top 5 teams which has the highest Winning percentage
dataFrame.sort_values(by='Winning Percent', ascending=False).head()
```

```python
#Group teams based on their “Best position” and print the sum of their points for all positions
dataFrame.groupby("BestPosition")['Points'].sum()
```
